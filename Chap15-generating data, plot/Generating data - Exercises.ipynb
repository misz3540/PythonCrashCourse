# 15-1 Cubes

```python
import matplotlib.pyplot as plt

%matplotlib inline
x_values = list(range(1,6))
cubes = [x**3 for x in x_values]
plt.plot(x_values, cubes, linewidth=5)

plt.title('Cubes', fontsize=24)
plt.xlabel('x', fontsize=14)
plt.ylabel('x^3', fontsize=14)

plt.tick_params(axis='both', labelsize=14)
plt.show()
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
x_values = list(range(1,5000))
cubes = [x**3 for x in x_values]
plt.plot(x_values, cubes, linewidth=5)

plt.title('Cubes', fontsize=24)
plt.xlabel('x', fontsize=14)
plt.ylabel('x^3', fontsize=14)

plt.tick_params(axis='both', labelsize=14)
plt.axis([0, 5100, 0, 5150**3])
plt.show()
```

# 15-2 Colored cubes

```python
import matplotlib.pyplot as plt

x_values = list(range(1, 5001))
y_values = [x**3 for x in x_values]

plt.scatter(x_values, y_values, c=y_values, cmap=plt.cm.twilight, edgecolor='none', s=40)

plt.title('Cubes', fontsize=24)
plt.xlabel('x', fontsize=14)
plt.ylabel('x^3', fontsize=14)

plt.tick_params(axis='both', labelsize=14)
plt.axis([0, 5100, 0, 5150**3])
plt.savefig('cubes_plot.png', bbox_inches='tight')
plt.show()
```

# 15-3 Molecular motion

```python
#rw_visualy.py
import matplotlib.pyplot as plt
%matplotlib inline
from random_walk import RandomWalk

while True:
    rw = RandomWalk(5000)
    rw.fill_walk()
    
    #Set the size of the plotting window
    plt.figure(dpi=128, figsize=(4, 3))
    #Plot the points, and show the plot
    point_numbers = list(range(rw.num_points))
    plt.plot(rw.x_values, rw.y_values, color='green', linewidth=3)
    plt.scatter(0, 0, c='blue', edgecolor='none', s=40)
    plt.scatter(rw.x_values[-1], rw.y_values[-1], c='red', edgecolor='none', s=40)
    # Remove axes
    plt.axes().get_xaxis().set_visible(False)
    plt.axes().get_yaxis().set_visible(False)
    
    plt.show()
    
    keep_running = input('Show another random walk plot? (y/n) ')
    if keep_running == 'n':
        break
    
    
```

# 15-6 Automatic labels

```python
# die_visual.py

import pygal
from die import Die

#Create a D6
die = Die()

# Make some rolls, and store results in a list.
results = []
for roll_num in range(1000):
    result = die.roll()
    results.append(result)
# print(results)

# Analyze the results
frequencies = []
for value in range(1, die.num_sides + 1):
    frequency = results.count(value)
    frequencies.append(frequency)
    
# print(frequencies)

# Visualize the results
hist = pygal.Bar()

hist.title = "Results of rolling one D6 1000 times."
hist.x_labels = ["'" + str(x) + "'" for x in range(1, die.num_sides + 1)] #['1', '2', '3', '4', '5', '6']
hist.x_title = "Result"
hist.y_title = "Frequency of result"

hist.add('D6', frequencies)
hist.render_to_file('die_visual.svg')
```

# 15-10 Practicing with both libraries

```python
import matplotlib.pyplot as plt

from die import Die

%matplotlib inline

#Create two D6 dice
die_1 = Die()
die_2 = Die()

# Make some rolls, and store results in a list.
results = []
for roll_num in range(2**16):
    result = die_1.roll() + die_2.roll()
    results.append(result)
# print(results)

# Analyze the results
frequencies = []
max_result = die_1.num_sides + die_2.num_sides
for value in range(2, max_result + 1):
    frequency = results.count(value)
    frequencies.append(frequency)
    
# print(frequencies)

# Visualize the results
plt.figure(dpi=2**7, figsize=(6, 5))
plt.bar(range(2, max_result + 1), frequencies)
plt.xlabel('sum of two D6s', fontsize=14)
plt.ylabel('frequencies', fontsize=14)

plt.tick_params(axis='both', labelsize=14)
plt.show()
```
