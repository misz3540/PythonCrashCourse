# 16-2 Sitka-Death Valley comparison

```python
from matplotlib import pyplot as plt
%matplotlib inline
import csv
from datetime import datetime

# Get dates and high temperatures from file
filename = 'pcc-master/chapter_16/death_valley_2014.csv'
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)
    
    dates, highs, lows = [], [], []
    for row in reader:
        try:
            current_date = datetime.strptime(row[0], "%Y-%m-%d")
            high = int(row[1])
            low = int(row[3])
        except ValueError:
            print(current_date, 'missing data')
        else:
            dates.append(current_date)
            highs.append(high)
            lows.append(low)
        
#     print(highs)
filename = 'pcc-master/chapter_16/sitka_weather_2014.csv'
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)
    
    dates2, highs2, lows2 = [], [], []
    for row in reader:
        try:
            current_date = datetime.strptime(row[0], "%Y-%m-%d")
            high = int(row[1])
            low = int(row[3])
        except ValueError:
            print(current_date, 'missing data')
        else:
            dates2.append(current_date)
            highs2.append(high)
            lows2.append(low)

# Plot data
fig = plt.figure(dpi=128, figsize=(10,6))
plt.plot(dates, highs, c='blue', alpha=0.5) #alpha controls a color's transparency. 0:transparent, 1:opaque
plt.plot(dates, lows, c='blue', alpha=0.5)
plt.fill_between(dates, highs, lows, facecolor='blue', alpha=0.1, label='Death Valley')
plt.legend()

plt.plot(dates2, highs2, c='red', alpha=0.5) #alpha controls a color's transparency. 0:transparent, 1:opaque
plt.plot(dates2, lows2, c='red', alpha=0.5)
plt.fill_between(dates2, highs2, lows2, facecolor='red', alpha=0.1, label='Sitka')
plt.legend()
# Format plot.
title = 'Daily high and low temperatures - 2014 \nDeath Valley, CA, vs. Sitka, AK'
plt.title(title, fontsize=20)
plt.xlabel('', fontsize=16)
fig.autofmt_xdate()
plt.ylabel("Temperature (F)", fontsize=16)
plt.ylim(20, 120)
plt.tick_params(axis='both', which='major', labelsize=14)

plt.show()
```

```python
# world_population.py

import json
import pygal
import pygal_maps_world.maps 
from pygal.style import RotateStyle as RS, LightColorizedStyle as LCS

from country_codes import get_country_code

# Load the data into a list
filename = 'pcc-master/chapter_16/population_data.json'
with open(filename) as f:
    pop_data = json.load(f)
    
# Print the 2010 population for each country
# Build a dictionary of population data
cc_populations = {}
for pop_dict in pop_data:
    if pop_dict['Year'] == '2010':
        country = pop_dict['Country Name']
        population = int(float(pop_dict['Value']))
        code = get_country_code(country)
        if code:
            cc_populations[code] = population

# Group the countries into 3 population levels
cc_pops_1, cc_pops_2, cc_pops_3 = {}, {}, {}
for cc, pop in cc_populations.items():
    if pop < 10**7:
        cc_pops_1[cc] = pop
    elif pop < 10**9:
        cc_pops_2[cc] = pop
    else:
        cc_pops_3[cc] = pop
#See how many countries are in each level
print(len(cc_pops_1), len(cc_pops_2), len(cc_pops_3) )

# The color used here (#336699) mixes a bit of red (33), a little more green (66), 
# and even more blue (99) to give RotateStyle a light blue base color to work from.
wm_style = RS('#336699', base_style=LCS) 
wm = pygal_maps_world.maps.World(style=wm_style)
wm.title = 'World population in 2010, by country'
wm.add('0-10m', cc_pops_1)
wm.add('10m-1bn', cc_pops_2)
wm.add('>1bn', cc_pops_3)

wm.render_to_file('world_population2.svg')
```

# 16-6 Gross domestic product

```python
# world_gdp.py

import json
import pygal
from pygal.maps.world import World
from pygal.style import RotateStyle as RS, LightColorizedStyle as LCS

from country_codes import get_country_code

# Load the data into a list
filename = 'global_gdp.json'
with open(filename) as f:
    gdp_data = json.load(f)
    
# Print the 'Years' found in the data set
years = []
for gdp_dict in gdp_data:
    if int(gdp_dict['Year']) in years:
        continue
    else:
        years.append(int(gdp_dict['Year']))
recent_year = max(years)
# print(gdp_dict)

# Print the recent year gdp for each country
# Build a dictionary of gdp data
cc_gdp = {}
for gdp_dict in gdp_data:
    if gdp_dict['Year'] == str(recent_year) or gdp_dict['Year'] == recent_year:
        country = gdp_dict['Country Name']
        gdp = int(float(gdp_dict['Value']))
        code = get_country_code(country)
        if code:
            cc_gdp[code] = gdp/10**0 # in million $s
#         else:
#             print('ERROR - ' + country)
print(min(cc_gdp.values())) # in million $s
print(max(cc_gdp.values())) # in million $s

# Group the countries into 3 population levels
cc_gdp_1, cc_gdp_2, cc_gdp_3 = {}, {}, {}
for cc, gdp in cc_gdp.items():
    if gdp < 10**9:
        cc_gdp_1[cc] = gdp
    elif gdp < 10**12:
        cc_gdp_2[cc] = gdp
    else:
        cc_gdp_3[cc] = gdp
#See how many countries are in each level
print(len(cc_gdp_1), len(cc_gdp_2), len(cc_gdp_3) )

# The color used here (#336699) mixes a bit of red (33), a little more green (66), 
# and even more blue (99) to give RotateStyle a light blue base color to work from.
wm_style = RS('#336699', base_style=LCS) 
wm = World(style=wm_style)
wm.force_uri_protocol = 'http'
wm.title = 'World gdp in ' + str(recent_year) + ', by country'
wm.add('0-1billion $', cc_gdp_1)
wm.add('1billion-1trillion $', cc_gdp_2)
wm.add('>1trillion $', cc_gdp_3)

wm.render_to_file('world_gdp.svg')
   
```

```python

import csv
import requests
from pygal.maps.world import COUNTRIES, World
from country_codes import get_country_code
from datetime import datetime, date

from pygal.style import RotateStyle as RS, LightColorizedStyle as LCS

CSV_URL = 'https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv'


with requests.Session() as s:
    download = s.get(CSV_URL)

    decoded_content = download.content.decode('utf-8')

    cr = csv.reader(decoded_content.splitlines(), delimiter=',')
    
    
    header_row = next(cr)
#     print(header_row)
countries = []
cc_last_update = {}
cc_total_cases = {}
cc_countries = {}
for row in cr:
    country = row[2]
    code = get_country_code(country)
    if code:
        if row[2] not in countries:
            countries.append(country)
            cc_countries[code] = row[2]
            cc_last_update[code] = datetime.strptime(row[3], "%Y-%m-%d")
            try: 
                total_cases = int(float(row[4]))
            except ValueError:
                pass #print(code, ' ', cc_last_update[code], ' -- missing data')
            else:
                cc_total_cases[code] = int(float(row[4]))
#                 cc_countries[code] = row[2]
        else:
            if datetime.strptime(row[3], "%Y-%m-%d") > cc_last_update[code]:
                cc_last_update[code] = datetime.strptime(row[3], "%Y-%m-%d")
                try: 
                    total_cases = int(float(row[4]))
                except ValueError:
                    pass #print(code, ' ', cc_last_update[code], ' -- missing data')
                else:
                    cc_total_cases[code] = int(float(row[4]))
#                     cc_countries[code] = row[2]

        
# print(cc_total_cases)
# print(min(cc_total_cases.values()))
# print(max(cc_total_cases.values()))

# Group the countries into 3 total cases levels
cc_total_cases_1, cc_total_cases_2, cc_total_cases_3 = {}, {}, {}
for cc, cases in cc_total_cases.items():
    if cases < 10**3:
        cc_total_cases_1[cc] = cases
    elif cases < 10**6:
        cc_total_cases_2[cc] = cases
    else:
        cc_total_cases_3[cc] = cases
#See how many countries are in each level
print(len(cc_total_cases_1), len(cc_total_cases_2), len(cc_total_cases_3) )

# The color used here (#336699) mixes a bit of red (33), a little more green (66), 
# and even more blue (99) to give RotateStyle a light blue base color to work from.
wm_style = RS('#336699', base_style=LCS) 
wm = World(style=wm_style)
wm.force_uri_protocol = 'http'
wm.title = 'World COVID19 total cases as of Dec 2020' + ', by country'
wm.add('0-1000', cc_total_cases_1)
wm.add('1000-1m', cc_total_cases_2)
wm.add('>1m', cc_total_cases_3)

wm.render_to_file('world_covid19_cases.svg')
```
